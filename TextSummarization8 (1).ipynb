```python
from transformers import BartTokenizer, BartForConditionalGeneration

# Load pretrained BART model and tokenizer
model_name = "facebook/bart-large-cnn"
tokenizer = BartTokenizer.from_pretrained(model_name)
model = BartForConditionalGeneration.from_pretrained(model_name)

# Input text
text = """
The Government of India on Monday announced a new national artificial intelligence initiative aimed at accelerating the country’s digital economy and improving public services. The program focuses on promoting AI research, supporting startups, and integrating artificial intelligence into key sectors such as healthcare, education, agriculture, and transportation.

According to officials, the initiative will provide funding for AI research institutions and encourage collaboration between universities, private companies, and government agencies. Special emphasis will be placed on ethical AI development, data privacy, and responsible use of emerging technologies.

The government also plans to introduce AI-based solutions in public healthcare systems to improve early disease detection and reduce the burden on medical professionals. In the education sector, AI-powered tools will be used to personalize learning experiences and identify skill gaps among students.

Industry experts welcomed the move, stating that the initiative could help India become a global leader in artificial intelligence. However, they also stressed the importance of upskilling the workforce to ensure that employees are prepared for AI-driven transformations.

The initiative is expected to be rolled out in phases over the next five years, with regular evaluations to measure its impact on economic growth and employment.
"""

# Tokenize input
inputs = tokenizer(
    text,
    max_length=1024,
    return_tensors="pt",
    truncation=True
)

# Generate summary
summary_ids = model.generate(
    inputs["input_ids"],
    max_length=80,
    min_length=30,
    length_penalty=2.0,
    num_beams=4,
    early_stopping=True
)

# Decode summary
summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)


```

```python
#comparison from original text to generated summary

print("Original Text:\n", text)
print("\nGenerated Summary:\n", summary)
```

```python


from rouge_score.rouge_scorer import RougeScorer

reference_summary = "India has announced a new national AI initiative to accelerate its digital economy and improve public services. The program focuses on promoting AI research, supporting startups, and integrating AI into sectors like healthcare, education, agriculture, and transportation. It will provide funding for AI research institutions and emphasize ethical AI development, data privacy, and responsible use. The initiative also plans AI-based solutions for public healthcare and personalized learning, with a phased rollout over five years."
scorer = RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
scores = scorer.score(reference_summary, summary)
print("ROUGE Scores:")
for rouge_type, score_tuple in scores.items():
    print(f"{rouge_type.upper()}:\n  Precision: {score_tuple.precision:.4f}\n  Recall: {score_tuple.recall:.4f}\n  F-measure: {score_tuple.fmeasure:.4f}")
```
